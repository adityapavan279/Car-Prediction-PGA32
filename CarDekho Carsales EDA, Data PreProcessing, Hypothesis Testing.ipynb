###  Hypothesis testing

```python
# Test Null Average selling_price of different transmission is equal
# using groupby
np.round(carsales.selling_price.groupby(carsales.transmission).mean(),2)
# since exactly 2 levels/groups/classes - 2sample independent T test
```

```python
np.round(carsales.selling_price.groupby(carsales.transmission).var(),2)
```

```python
# Frame Null and Alternate Hypothesis

# Null- there is no significant difference in averagr selling_price of auto and manual cars
# H0 - mean of auto = mean of manual

# Alternate -there is significant difference in averagr selling_price of auto and manual cars
# Ha - mean of auto != mean of manual
```

```python
# Split Data into Auto and Manual
auto = carsales[carsales.transmission == "Automatic"]
manual = carsales[carsales.transmission == "Manual"]
```

```python
pip install scipy
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(auto.selling_price,manual.selling_price,equal_var=False)
# since pvalue=8.531946079710315e-129 is less than 0.05, Reject Null
# Difference in Means is statistically significant
```

```python
# The distribution of dependent variable decides whether to use Regression or Classification algorithim

# Regression Algorithim - Dependent Variable(y) is numerical, continuos, closer to normal distribution(bell curve)
# with minimum outliers and no missing values
```

```python
# Classification Algorithim- Dependent variable is non numeric either binary(yes/no) ormultinomial more than
#levels/classes/groups

# Multiple Linear Regression(MLR) explains linear releationship between dependent variable(y) and multiple independet vaariables(x's)
# MLR is an equation
# y = B0+B1X1+B2X2+B3X3+...........+BnXn+Et
# y == Dependent variable or prediction variable
# B0 - Intercept or constant(point at which straight lline touched y axis)
# B1,B2,B3,..............Bn - Coefficients
# X1,X2,X3,X4..............Xn - Independent variables or predictors
# Et  - Error Term or residual  = Actualvalue - Predict value


# ASSUMPTIONS - MLR is a parametric algorithm(many assumptions)   PRE MODEL ASSUMPTIONS
# 1) Depedent Variable y must be numerical,continuos,closer to normal distributionwith minimum outliers and no missing values
# 2) Independent variable(X's) can be both numerical and numericalcategorical
# 3) No Multicollinearity - Multicollinearity is very strong correlation (>0.90) among variables. If Multicollinearity is found do not include both 
#    variables and choose one among them, else overfitting will occur.
# 4) Exogentity - Dependent variable is dependent on independent variables but not viceversa.
#   y is dependent on X's  but X's do not depend on y
# 5) Varience Inflation Factor(VIF) - VIF Greater than 5 indicates  multicollinearity
# 6) There must be a logical linear relationship between dependent variable and independent variables

# POST MODEL ASSUMPTIONS
#7) Residuals must be normally distibuted. Check during Q-Q plot.
# 8) residuals must be homoscedasity in nature



# The Interpretation of output (python only)
# R Square must be between 0.60 - 0.95. R Square explains amount of varience occuring in dependent variable caused by independent variables or model
# less than 0.60 - underfitting
# greater than 0.95 - overfitting

# Intercept + Coefficients for creating predictive equation

# Statistical Interpretaion of output
# 1) Adjusted R Square mst be between 0.60 - 0.95. Adj R Square is sensitive to independent variable's siginificance, if independent variable is 
# siginificant it will increase and if variable is insiginificant it will decrease.

# 2) Model P-Values(lies below Adj R Square) or significance F in MS Excel must be less than 0.05
# Null - Coefficients(B's) are equal to Zero.
# Alt - Coefficients(B's) are not equal to zero.

# 3) Independent Variables(X's) p-value must be less than 0.5, Reject Null
# Null - Independent Variable is Insignificant
# Alt - Independent Variable is Significant
```
